```python
import os
os.makedirs('video-frame', exist_ok=True)
os.makedirs('video-frame-detection', exist_ok=True)
```

```python
import cv2
vidcap = cv2.VideoCapture('video.mp4')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite("video-frame/%06d.jpg" % count, image)     # save frame as JPEG file
  success,image = vidcap.read()
  if success is False:
    continue
    
  count += 1

print("finish! convert video to frame")
```

```python
from natsort import natsorted
import os.path as osp
from ultralytics import YOLO

model = YOLO('yolov8/runs/train/weights/best.pt')
blue_color = (255, 0, 0)

origin_suffix = 'video-frame'
origin = os.listdir(origin_suffix)
origin = natsorted(origin) ## order by desc

cnt = 10

for f in origin:
    img = cv2.imread(osp.join(origin_suffix, f))
    
    results = model.predict(img)
    for r in results:
        boxes = r.boxes
        for box in boxes:
            b = box.xyxy[0]
            c = box.cls
            
            x1, y1, x2, y2 = int(b[0]), int(b[1]), int(b[2]), int(b[3])

            img = cv2.rectangle(img, (x1, y1), (x2, y2), blue_color, 3)
            img = cv2.putText(
                img,
                "Fire",
                (x1, y1 - 15 if y1 - 15 > 15 else y1 + 15),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                blue_color,
                2,
            )
    
    cv2.imwrite(f'video-frame-detection/{cnt}.png', img)
    cnt += 1
```

```python
import os
import moviepy.video.io.ImageSequenceClip
image_folder='video-frame-detection'
fps=30

image_files = os.listdir('video-frame-detection')
image_files = natsorted(image_files)
image_files = [osp.join('video-frame-detection', img) for img in image_files]

clip = moviepy.video.io.ImageSequenceClip.ImageSequenceClip(image_files, fps=fps)
clip.write_videofile('result.mp4')
```
