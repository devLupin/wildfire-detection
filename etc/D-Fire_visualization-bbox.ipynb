```python
import os
from PIL import Image, ImageDraw
import csv
import matplotlib.pyplot as plt
```

```python
def yolobbox2bbox(x, y, w, h, width, height):
    x1, y1 = x-w/2, y-h/2
    x2, y2 = x+w/2, y+h/2
    return x1*width, y1*height, x2*width, y2*height
```

```python
img_path = 'train/images\WEB09436.jpg'
label_path = 'train/labels\WEB09436.txt'
```

```python
img = Image.open(img_path).convert('RGB')
draw = ImageDraw.Draw(img)
width, height = img.size
```

```python
def yolobbox2bbox(x, y, w, h, width, height):
    x1, y1 = x-w/2, y-h/2
    x2, y2 = x+w/2, y+h/2
    return x1*width, y1*height, x2*width, y2*height
```

```python
f = open(label_path, 'r')
info = f.readlines()

for cur in info:
    cur = cur[:-1]
    cls, x, y, w, h = cur.split(' ')
    
    
    
    x, y, w, h = float(x), float(y), float(w), float(h)
    x, y, w, h = yolobbox2bbox(x, y, w, h, width, height)
    print(cls, x, y, w, h)
    draw.rectangle((x, y, w, h), outline=(0,255,0), width=3)
    
plt.imshow(img)
plt.show()
```
